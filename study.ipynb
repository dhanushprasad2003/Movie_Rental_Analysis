```python


# 1. Load our CSV files into Pandas DataFrames (think of these as Excel sheets)
# Your files are inside the 'movie_dataset' folder, so we add that to the path!
rentals = pd.read_csv('movie_dataset/rental.csv')
customers = pd.read_csv('movie_dataset/customer.csv')
payments = pd.read_csv('movie_dataset/payment.csv')

# 2. Check the data types of our Rentals table
print("Before cleaning:")
print(rentals.dtypes)

# 3. Clean our dates! Let's find any column with 'date' in its name and convert it.
# We will use the rentals table as an example.
for col in rentals.columns:
    if 'date' in col.lower():
        # Convert to datetime, coercing errors to 'NaT' (Not a Time) so we don't crash
        rentals[col] = pd.to_datetime(rentals[col], errors='coerce')

print("\nAfter cleaning:")
print(rentals.dtypes)
```

```python
# 1. Total Revenue (adding up all the payments)
total_revenue = payments['amount'].sum()

# 2. Total Customers (counting unique people)
total_customers = customers['customer_id'].nunique()

# 3. Average Order Value (the average amount per payment)
average_order_value = payments['amount'].mean()

# 4. Let's print them out with some nice formatting!
print(f"💰 Total Revenue: ${total_revenue/1000:.2f}k")
print(f"👥 Total Customers: {total_customers:,}")
print(f"💳 Average Order Value: ${average_order_value:,.2f}")
```

```python
# 1. Merge (VLOOKUP) our payments and customers tables together.
# They both share the 'customer_id' column, so we use that as the bridge to connect them!
customer_spending = pd.merge(payments, customers, on='customer_id', how='left')

# 2. Group by (Pivot Table) to find our best customers!
top_customers = (
    customer_spending.groupby(['first_name', 'last_name'])['amount']
    .sum()
    .reset_index()
    .sort_values(by='amount', ascending=False)
)

print("🏆 Top 5 Customers by Revenue:")
print(top_customers.head(5))
```
